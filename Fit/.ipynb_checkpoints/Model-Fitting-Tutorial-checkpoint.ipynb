```python
import numpy as np
import numpy.matlib
import optimization as opt
import matplotlib.pyplot as plt
```

# Model Fitting

In this tutorial we'll cover how to fit a model to data using SciPy's minimization routine [`scipy.optimize.fmin`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fmin.html). Model fitting is a procedure that takes three steps:

1. You need a function that takes in a set of parameters and returns a predicted data set.

2. You need an 'error function' that provides a number representing the difference between your data and the model's prediction for any given set of model parameters. This is usually either the sums of squared error (SSE) or maximum likelihood.

3. You need to find the parameters that minimize this difference.

## One Parameter Example - Weber's Law

We'll start with a simple example where our model has just one parameter. Weber's law states that the ability for a subject to notice an increase in stimulus intensity is proportional to the starting, or baseline intensity. That is, if $x$ is the stimulus intensity, the increment threshold is $k_{x}$, where $k$ is the 'Weber fraction'. This fraction is our one parameter.

Suppose that we ran an experiment testing the ability for a subject to detect an increase in the weight of an object held in the hand. Subjects were given starting weights (in Kg) of the following values:

```python
x = np.arange(0.5, 3.5, 0.5)
print(x)
```

Increment thresholds are defined here as the increase in weight that can be detected correctly 80% of the time. This would use a psychophysical method such as 'two-alternative forced-choice' (2AFC) that we don't need to deal with right here. Let the corresponding increment thresholds for a subject be:

```python
y = np.array([0.0619, 0.0888, 0.1564, 0.1940, 0.2555, 0.2890])
print(y)
```

We can plot the increment threshold `y` as a function of the baseline weight `x`.

```python
fig = plt.figure()
ax = plt.scatter(x, y, c='black')
plt.xlabel('Baseline Weight (Kg)')
plt.ylabel('Increment Threshold (Kg)')
plt.show()
```

## Step 1: Model Prediction

We're now ready for the first step - writing the function containing the model that predicts the data. This function needs to take in a single parameter and the baseline weights and return a prediction of the data. The parameter is the Weber Fraction which is the slope of the line of the data in figure 1.

We'll use a specific convention for how we represent our parameters which is to place them inside a single dictionary. With scalar parameter this seems a little silly, but it'll make sense when we add more parameters.

Here's a dictionary containing a starting guess at the Weber fraction `w` and the baseline values `x`:

```python
params = {'w': 0.09, 'x': x}
print(params)
```

Our function `WebersLaw` is a single line function that will take in the Weber Fraction as its first argument, and the list of baseline values as the second argument:

```python
def WebersLaw(w,x):
    return(w*x)
```

We can pass the our parameter dictionary with keys that match the key worded arguments in our `WebersLaw` function with some Python syntax:

```python
pred = WebersLaw(**params)
```

and then plot the prediction

```python
plt.scatter(x, y, c='black', label='Data')
plt.plot(x, pred, c='red', label='First Guess')
plt.xlabel('Baseline Weight (Kg)')
plt.ylabel('Increment Threshold (Kg)')
plt.legend(loc='upper left', fontsize='medium')
plt.show()
```

## Step 2: Comparing the Model Prediction to the Data

Time for Step 2. Here's an error function `WebersLawErr` that calculates the sums of squared error between the model's prediction and the data:

```python
def WebersLawErr(w,x,y):
    return(np.sum((WebersLaw(w,x) - y)**2))
```

```python
err = WebersLawErr(**params, y=y)
print('\nThe SSE between our model and the data from our first guess is:\n\n\t%5.4f' % err)
```

## Step 3: Finding the Best Fitting Parameter

What is the best fitting Weber fraction? With one parameter we can visualize this problem by calculating the error of the fits for a range of Weber fractions:

```python
WeberList = np.linspace(0.08, 0.12, 41)
errList = [WebersLawErr(i,x,y) for i in WeberList]

fig = plt.figure()
plt.plot(WeberList, errList, c='black')
plt.xlabel('Weber Fraction')
plt.ylabel('SSE')
plt.show()
```

We can plot the error value of our initial guess in red:

```python
fig = plt.figure()
plt.plot(WeberList, errList, c='black', zorder=1)
plt.scatter(params['w'], err, c='red', edgecolors='red', zorder=2)
plt.xlim((min(WeberList), max(WeberList)))
plt.ylim((0, 0.012))
plt.xlabel('Weber Fraction')
plt.ylabel('SSE')
plt.show()
```

It looks like the best fitting Weber fraction is between 0.095 and 0.1. We could go further and keep sampling finer and finer in this range to find a minimum value. But this is a very inefficient way to find the minimum of a function. It's especially bad if there are multiple parameters to fiddle with.

Fortunately, SciPy has a function [`scipy.optimize.fmin`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fmin.html) that uses a sophisticated numerical analysis technique that can find the minimum of a function like this - as long as it's reasonably continuous and well behaved. SciPy's implementation of this non-linear optimization search to be a bit kludgy and awkward, so I've provided a wrapper class `fit` that calls [`scipy.optimize.fmin`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fmin.html) with a method `.fit()` that I find easier to use. The class is defined in [`optimization.py`](https://github.com/kellychang4/Tutorials/blob/master/Fit/optimization.py) and it is called as below:

1. Creating a dictionary with keys that match ALL the keyworded argument names of the provided error function (`p`)
2. Creating a dictionary with key(s) that match the parameters names that will be free to vary while fitting (`freeList`). The value for the free parameters specify the constaints applied to the optimization. For this example there are no constaints, so we supply the `None` for no constraints.
3. Initializing the `fit` object with error function, initial parameters, and free parameter dictionary
4. Calling the `.fit()` method on the `fit` object

```python
p = {'w': 0.09, 'x': x, 'y': y} 
freeList = {'w': None} 
model = opt.fit(func=WebersLawErr, params=p, freeList=freeList)
print(model) # ignore the mask attribute for this example
```

```python
[pBest,errBest] = model.fit()
```

```python
print(pBest['w'])
```

We can plot this in Figure 2

```python
fig = plt.figure()
plt.plot(WeberList, errList, c='black', zorder=1)
plt.scatter(p['w'], err, c='red', edgecolors='red', zorder=2, label='Initial Guess')
plt.scatter(pBest['w'], errBest, c='blue', edgecolors='blue', zorder=3, label='Fitted')
plt.xlim((min(WeberList), max(WeberList)))
plt.ylim((0, 0.012))
plt.xlabel('Weber Fraction')
plt.ylabel('SSE')
plt.legend(loc='upper left', fontsize='medium')
plt.show()
```

See how it sits on the bottom of the curve? We found the optimal Weber fraction that fits our data. About 0.1033 (or 10.3%)

Finally, we'll get the model prediction using this best fitting value and plot the predictions with the data in Figure 1

```python
predBest = WebersLaw(pBest['w'], pBest['x'])

plt.scatter(x, y, c='black', label='Data')
plt.plot(x, pred, c='red', label='First Guess')
plt.plot(x, predBest, c='blue', label='Best Value')
plt.xlabel('Baseline Weight (Kg)')
plt.ylabel('Increment Threshold (Kg)')
plt.legend(loc='upper left', fontsize='medium')
plt.show()
```

## Other Measures of Error

The best fitting parameters can depend strongly on your choice of error function. For our example, another way to state Weber's law is that the ratio of the increment thresholds and the baseline values is constant. So another measure of the Weber fraction is the mean of these ratios across baselines:

```python
meanRatio = np.mean(y/x)
print('\n\tBest w: %5.4f\n\tMean Ratio: %5.4f\n' % (pBest['w'], meanRatio))
```

This difference is at least in part caused by a greater weight for the mean ratio on the lower baseline values (or equivalently a greater weight for the high baseline values for the model). It's not obvious which is more valid. The model puts equal weight on absolute errors in the `y` dimension across baselines, while the mean ratio puts equal weights on the ratios across baselines.

## Model Fitting Weighting by Standard Error of the Mean

The mean ratios might be a better method because the variability in the data (which we haven't yet talked about) probably increases with the baseline. So errors at high baselines are less meaningful than at low baselines.

A natural way to correct for variance in the data is to normalize the SSE by the standard error of the mean for each data point. Then our error value will represent a total error in sqsuared standard error units.

Let the standard error of the mean of our estimates be:

```python
s = np.array([0.016, 0.0216, 0.0140, 0.0225, 0.0183, 0.0517])

plt.errorbar(x, y, c='black', yerr=s, fmt='o', label='Data')
plt.plot(x, pred, c='red', label='First Guess')
plt.plot(x, predBest, c='blue', label='Best Value')
plt.xlim((0, 3.5))
plt.xlabel('Baseline Weight (Kg)')
plt.ylabel('Increment Threshold (Kg)')
plt.legend(loc='upper left', fontsize='medium')
plt.show()
```

```python
def WebersLawErr(w,x,y,s=np.ones(np.size(y))):
    return(np.sum(((WebersLaw(w,x)-y)**2)/(s**2)))
```

Let's re-fit our new `WebersLawErr` function. We'll need to update our parameter dictionary`p` with a new `s` key and values. But once that's done, we can just recall the `.fit()` method with our new function `func` and paramater dictionary. The `fit` object `model` all the variables it was initalized with, so our `freeList` from before is still there.

```python
p.update({'s': s}) 
[pBestSEM,errBestSEM] = model.fit(func=WebersLawErr, params=p)
predBestSEM = WebersLaw(pBestSEM['w'], x)
```

```python
plt.errorbar(x, y, c='black', yerr=s, fmt='o', label='Data')
plt.plot(x, pred, c='red', label='First Guess')
plt.plot(x, predBest, c='blue', label='Best fit using SSE')
plt.plot(x, predBestSEM, c='black', label='Best fit using SEM')
plt.xlim((0, 3.5))
plt.xlabel('Baseline Weight (Kg)')
plt.ylabel('Increment Threshold (Kg)')
plt.legend(loc='upper left', fontsize='medium')
plt.show()
```

The new best fit takes into account the error bars so that data points with smaller error bars are weighted more heavily. You can think of this as making the model prediction overlap more with the error bars, rather than simply trying to get close to the data points.

In this example, the last data point (highest baseline value) is noisy, so the model's constraint to fit this point is relaxed. This causes the best fitting Weber fraction to change so that the prediction moves away from this noisy data point.

## Model Fitting Weighting by Indivdual Measurements

Another way to account for the variability in your measurements is to use an error function that sums up the errors with respect to each individual measurements, rather than the means. This is closely related to the previous method - it's identical in the limit as you add more and more measurements to each mean.

We'll start with a new sample data set from the Weber's law experiment that has four measurements for each baseline intensity. We'll call the variable ``yy`` instead of `y`.

```python
yy = np.array([[0.0457, 0.0885, 0.1533, 0.1941, 0.2607, 0.3029], 
               [0.0333, 0.1119, 0.1517, 0.2218, 0.2506, 0.2866], 
               [0.0513, 0.1119, 0.1481, 0.1986, 0.2490, 0.3071], 
               [0.0529, 0.0996, 0.1573, 0.2011, 0.2417, 0.3162]])
```

Each of the 4 rows corresponds to a separate measurement across the 6 baseline values.

An easy way to compare these values to the model's prediction is to make the model predict a data set that's the same size as `yy`. This can be done by using a matrix ``xx`` that's the same size too:

```python
xx = np.matlib.repmat(x,4,1)
print(xx)
```

We can plot this new data set as individual data points:

```python
plt.scatter(xx, yy, c='None')
plt.xlabel('Baseline Weight (Kg)')
plt.ylabel('Increment Threshold (Kg)')
plt.show()
```

It turns out that we've written our model function 'WebersLaw' so that it can take in matrices as well as vectors, so it will also make a predicted data set that's a matrix:

```python
w = 0.12
pred = WebersLaw(w,xx)
print(pred)
```

We've also written our error function `WebersLawErr` so that it can calculates the SSE between the elements of two matrices as well as vectors:

```python
err = WebersLawErr(w,xx,yy)
print(err)
```

This means that our fitting routine will work too, just create a new the parameter dictionary (or update, if you prefer):

```python
p = {'w': w, 'x': xx, 'y': yy}
[pBest,errBest] = model.fit(params=p)
pred = WebersLaw(pBest['w'], x)
```

```python
plt.scatter(xx, yy, c='None', label='Data')
plt.plot(x, pred, c='black', label='Best fitting model')
plt.xlabel('Baseline Weight (Kg)')
plt.ylabel('Increment Threshold (Kg)')
plt.legend(loc='upper left', fontsize='medium')
plt.show()
```

## Model Fitting with More Than One Parameter

Fitting a model that has more than one parameter is easy, since the hard part of actually finding the best parameters is all done by SciPy's [`scipy.optimize.fmin`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fmin.html) method. Here's an example of a data set that needs a two-parameter model to fit it.

Suppose we're measuring the firing rate of a neuron while it is recovering from an adapted state. Let `t` be time in seconds, and `y` be the firing rate of the neuron. Here's an example data set:

```python
t = np.arange(0,11)
y = np.array([0.4346, 9.8079, 14.2634, 18.2656, 23.4209, 27.8469,
              27.5358, 29.4886, 26.4415, 28.2591, 32.8832])
```

Plot the data

```python
plt.scatter(t, y, c='black', edgecolors='black')
plt.xlim((min(t)-0.1, max(t)+0.1))
plt.xticks(t)
plt.xlabel('Time (min)')
plt.ylabel('Firing Rate (Hz)')
plt.show()
```

Notice how the recovery from adaptation starts out quickly and then asymptotes to a firing rate of about 30Hz. A standard model for this sort of asymptotic function is an exponential. This can be described with two parameters - one describing the asymptotic firing rate (we'll call it `rMax`) and a second one decribing the rate of recovery (we'll call it `k`).

We need to do Steps 1 and 2 now from above - write the function containing the model `predRecovery` and write a function comparing the model to the data `predRecoveryErr`. 

```python
def predRecovery(rMax,k,t):
    return(rMax*(1-np.exp(-k*t)))

def predRecoveryErr(rMax,k,t,y,s=np.ones(np.size(y))):
    return(np.sum(((predRecovery(rMax,k,t)-y)**2)/(s**2)))
```

Notice that the model itself is just one line again: 

$$rMax * (1-e^{-kt})$$

We're ready to use this function to predict and fit data. 

Initializing the inital parameters:

```python
params = {'rMax': 30, 'k': 0.25}
```

Let's get fancy and use a higher sampling rate in time for the model prediction to get a smoother curve. Instead of predicting with the variable `t`, we'll use `tPlot`:

```python
tPlot = np.linspace(-0.9,10.1,101)

pred = predRecovery(**params, t=tPlot)

plt.scatter(t, y, c='black', edgecolors='black', zorder=2)
plt.plot(tPlot, pred, c='red', zorder=1)
plt.xlim((min(t)-0.1, max(t)+0.1))
plt.xticks(t)
plt.xlabel('Time (min)')
plt.ylabel('Firing Rate (Hz)')
plt.show()
```

Intial error value:

```python
err = predRecoveryErr(**params, t=t, y=y)
print(err)
```

Find the best fitting parameters:

Our call to '.fit()' will be just as in the one parameter case, except that we'll list two parameters in `freeList` dictionary:

```python
params.update({'t': t, 'y': y})
freeList = {'rMax': None, 'k': None}

model = opt.fit(func=predRecoveryErr, params=params, freeList=freeList)
[pBest,errBest] = model.fit()
```

Now we'll get the best fitting prediction and plot it in black

```python
predBest = predRecovery(pBest['rMax'], pBest['k'], tPlot)

plt.scatter(t, y, c='black', edgecolors='black', zorder=3, label='Data')
plt.plot(tPlot, pred, c='red', zorder=1, label='Initial Guess')
plt.plot(tPlot, predBest, c='black', zorder=2, label='Best Fit')
plt.xlim((min(t)-0.1, max(t)+0.1))
plt.ylim((0, 35))
plt.xticks(t)
plt.xlabel('Time (min)')
plt.ylabel('Firing Rate (Hz)')
plt.legend(loc='lower right', fontsize='medium')
plt.text(5, 15, ('rMax = %5.2f Hz\nk = %5.2f' % (pBest['rMax'], pBest['k'])))
plt.show()
```

## Holding Variables Constant while Fitting

The `.fit()` method lets you easily fit with a subset of parameters. Suppose that you know that the asymptotic firing rate is 30Hz, and we just want to let the parameter `k` vary. We do this by only listing `k` as a free parameter in `freeList`:

```python
[pkBest,pkErrBest] = model.fit(freeList={'k': None})
predKBest = predRecovery(pkBest['rMax'], pkBest['k'], tPlot)
```

```python
plt.scatter(t, y, c='black', edgecolors='black', zorder=4, label='Data')
plt.plot(tPlot, pred, c='red', zorder=1, label='Initial Guess')
plt.plot(tPlot, predBest, c='black', zorder=2, label='Best Fit')
plt.plot(tPlot, predKBest, c='blue', zorder=3, label=('Best w/ rMax = %d' % pkBest.get('rMax')))
plt.xlim((min(t)-0.1, max(t)+0.1))
plt.ylim((0, 35))
plt.xticks(t)
plt.xlabel('Time (min)')
plt.ylabel('Firing Rate (Hz)')
plt.legend(loc='lower right', fontsize='medium')
plt.text(5, 15, ('rMax = %5.2f Hz\nk = %5.2f' % (pBest.get('rMax'), pBest.get('k'))))
plt.show()
```
